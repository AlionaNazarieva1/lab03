```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
x0 = 33333
y0 = 44444
t0 = 0
a = 0.28
b = 0.745
c = 0.613
h = 0.35
tmax = 1.05
dt = 0.05
def p(t):
    p = abs(2*math.sin(3*t))
    return p
def q(t):
    q = abs(1.5*math.cos(2*t))
    return q
def dy(s,t):
    dy1 =  - a*s[0] - b*s[1] + p(t)
    dy2 = - c*s[0]*s[1] - h*s[1] + q(t)
    return [dy1, dy2]
t = np.arange(t0,tmax,dt) 
v0=[x0,y0]
s = odeint(dy,v0,t)
plt.plot(t,s[:,0],'r--', linewidth=2.0,label="численность x")
plt.plot(t,s[:,1],'b-', linewidth=2.0,label="численность y")
plt.xlabel("шаг")
plt.ylabel("численность армии")
plt.legend()
plt.grid()
plt.show()
```
